```python
import os
from PIL import Image
```

```python
# Set the directory containing your .tif images
input_dir = ".\Fingerprint-Feature-Extraction\DB1_A"   # change to your actual input path
output_dir = "./jpg_images"

# Create output directory if it doesn't exist
os.makedirs(output_dir, exist_ok=True)

# Number of main folders (e.g., 1, 2, 3...)
num_folders = 100  # You can change this based on your data
images_per_group = 8
```

```python
for group in range(1, num_folders + 1):
    for img_num in range(1, images_per_group + 1):
        tif_filename = f"{group}_{img_num}.tif"
        tif_path = os.path.join(input_dir, tif_filename)
        
        if not os.path.isfile(tif_path):
            print(f"File not found: {tif_path}")
            continue
        
        try:
            with Image.open(tif_path) as img:
                rgb_img = img.convert("RGB")
                jpg_filename = f"{group}_{img_num}.jpg"
                jpg_path = os.path.join(output_dir, jpg_filename)
                rgb_img.save(jpg_path, "JPEG")
                print(f"Converted: {tif_filename} -> {jpg_filename}")
        except Exception as e:
            print(f"Error converting {tif_filename}: {e}")
```
