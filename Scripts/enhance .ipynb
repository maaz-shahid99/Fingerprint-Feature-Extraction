```python
import os
import cv2
from PIL import Image
import numpy as np
```

```python
# Use raw string or os.path.join to avoid Windows path issues
input_dir = r"C:\Users\golut\OneDrive\Documents\Projects\Test demo 2\Fingerprint-Feature-Extraction\DB1_A"
output_dir = r"C:\Users\golut\OneDrive\Documents\Projects\Test demo 2\Fingerprint-Feature-Extraction\jpg_images"

os.makedirs(output_dir, exist_ok=True)

num_folders = 100
images_per_group = 8
```

```python
for group in range(1, num_folders + 1):
    for img_num in range(1, images_per_group + 1):
        tif_filename = f"{group}_{img_num}.tif"
        tif_path = os.path.join(input_dir, tif_filename)

        if not os.path.isfile(tif_path):
            print(f"File not found: {tif_path}")
            continue

        try:
            # Open with PIL and convert to grayscale
            pil_img = Image.open(tif_path).convert("L")

            # Convert to NumPy array for OpenCV
            img_np = np.array(pil_img)

            # Invert the image
            inverted = cv2.bitwise_not(img_np)

            # Enhance contrast using histogram equalization
            enhanced = cv2.equalizeHist(inverted)

            # Save as JPEG using OpenCV
            jpg_filename = f"{group}_{img_num}.jpg"
            jpg_path = os.path.join(output_dir, jpg_filename)
            cv2.imwrite(jpg_path, enhanced)

            print(f"Converted & enhanced: {tif_filename} -> {jpg_filename}")

        except Exception as e:
            print(f"Error converting {tif_filename}: {e}")
```
